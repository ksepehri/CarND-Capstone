### DBW Data Visualization

Requirements:
* pandas
* matplotlib
* bokeh (https://bokeh.pydata.org)

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()
```

```python
def read_dbw_data(filename):
    df = pd.read_csv('data/{}'.format(filename))
    #df=df[:2000]
    df['time'] = df['time']-df['time'][0]
    df['t_accel'] = df['t_speed']-df['c_speed']
    return df
    
```

```python
df = read_dbw_data('dbw_data.csv')

p = figure(title="acceleration", x_axis_label='time', y_axis_label='value')

p.line(df['time'], df['t_accel'], #/df.max()['t_accel'], 
       legend="t_acc", line_color="gray")
p.line(df['time'], df['throttle'], legend="throttle", line_color="red")
p.line(df['time'], df['c_speed'], legend="speed", line_color="blue")
# p.line(df['time'], df['brake'], legend="speed", line_color="yellow")

show(p)
```

```python
df = read_dbw_data('dbw_data.csv')

p = figure(title="steer", x_axis_label='time', y_axis_label='value')

p.line(df['time'], df['t_yd'], legend="t_yd", line_color="gray")
p.line(df['time'], df['steer']/14.8, legend="steer", line_color="red")
#p.line(df['time'], df['c_speed'], legend="speed", line_color="blue")
# p.line(df['time'], df['brake'], legend="speed", line_color="yellow")

show(p)

```
